# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

gb = df.groupby('neighbourhood_group') #.count() gets count of all cols
gb['id'].count() #counting num of ID columsn? Unsure if solution
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

# gb = df.groupby('neighbourhood_group')
gb['price'].mean() #finding mean price. Price is col name
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

# gb = df.groupby('neighbourhood_group')
gb['price'].agg(['min', 'max']) #use aggregate here to specify multiple operations. In our case, min and max are ops. Price is column name
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.

#double group by, get the two columns first
cols = ['neighbourhood_group', 'room_type'] #first, specify the cols we want to group by. Neighbourhood group and the room type. These are the column names
room_neigh = df.groupby(cols) #create new df with the group by of the TWO columns
room_neigh['price'].mean() #now find out the MEAN of the price (name of col) for each type of room in each neighbourhood group

# df.groupby(['neighbourhood_group', 'room_type'])['price'].mean() # one liner
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

agg = df.groupby('neighbourhood_group')
agg['price'].agg(['count','min','max','mean','median','std'])

# df.groupby('neighbourhood_group').price.agg(['count','min','max','mean','median','std']) #one liner also note we can write .price to reference column instead of writing ['price'] either one is good
```

# Use the grammys.csv data for the next section of questions.

1. Who was won Album of the Year in 2016?


2. Who won Best Rap Album in 2009?


3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?

```python
grammys = pd.read_csv('data/grammys.csv')
grammys.head() #just trying to see col names. We need double groupby so group by the year and the category
```

```python
#  1. Who was won Album of the Year in 2016?

con1 = grammys['category'] == 'Album of the Year' #double conditional. First condition, category is album of year
con2 = grammys['year'] == 2016 #second conditional, year is 2016. 
con3 = grammys['winner'] == True #third condition, we DONT care for nominees
grammys[con1 & con2 & con3] #result shows all cols
```

```python
# 2. Who won Best Rap Album in 2009?
#one liner
grammys[(grammys['year'] == 2009) & (grammys['category'] =='Best Rap Album') & (grammys['winner'] == 1)].workers
#essentially we do (condition1) & (condition2) & ...
#1 at the ends mean winner == True
# .workers prints the name of the column
```

```python
# 3. How many awards was Kendrick Lamar nomiated for, and how many did he win...?

print("Kendrick nominated for"
, grammys[(grammys['workers'] == 'Kendrick Lamar')]["nominee"].count() #here we are COUNTING the 'nominee' column where the worker name is Kendrick Lamar
, "grammys and he won" 
, grammys[(grammys['workers'] == 'Kendrick Lamar') & (grammys['winner'] == 1)]["winner"].count() 
)

```
